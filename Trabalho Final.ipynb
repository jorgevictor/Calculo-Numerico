# Métodos Diretos

### Eliminação Gaussiana



```python
def eliminacao_gaussiana(A, b):
    start = time.time()
    n = A.shape[0]
    x = np.zeros(n)

    a = np.zeros((n, n + 1))
    for i in range(n):
        a[:, i] = A[:, i]

    a[:, n] = b.T

    for i in range(n):
        if a[i][i] == 0.0:
            sys.exit('Divisão por zero!')

        for j in range(i + 1, n):
            ratio = a[j][i] / a[i][i]

            for k in range(n + 1):
                a[j][k] = a[j][k] - ratio * a[i][k]

    x[n - 1] = a[n - 1][n] / a[n - 1][n - 1]

    for i in range(n - 2, -1, -1):
        x[i] = a[i][n]

        for j in range(i + 1, n):
            x[i] = x[i] - a[i][j] * x[j]

        x[i] = x[i] / a[i][i]
    end = time.time()
    total = end - start
    return x, total

elimGauss = eliminacao_gaussiana(A, b)
print(elimGauss)
```

### Fatoração LU

```python
start = time.time()

def lu(A):
    n = A.shape[0]
    
    U = A.copy()
    L = np.eye(n, dtype=np.double)
    U = U.astype('float64')
    L = L.astype('float64')

    for i in range(n):

        factor = U[i+1:, i] / U[i, i]
        L[i+1:, i] = factor
        U[i+1:] -= factor[:, np.newaxis] * U[i]
 
    return L, U


def sub_1(L, b):

    n = L.shape[0]

    y = np.zeros_like(b, dtype=np.double);

    y[0] = b[0] / L[0, 0]

    for i in range(1, n):
        y[i] = (b[i] - np.dot(L[i,:i], y[:i])) / L[i,i]

    return y


def sub_2(U, y):


    n = U.shape[0]

    x = np.zeros_like(y, dtype=np.double);

    x[-1] = y[-1] / U[-1, -1]

    for i in range(n-2, -1, -1):
        x[i] = (y[i] - np.dot(U[i,i:], x[i:])) / U[i,i]

   
    return x


def lu_solve(A, b):

    start = time.time()
    L, U = lu(A)
    
    y = sub_1(L, b)
    
    end = time.time()
    total = end - start   
    return sub_2(U, y), total
    
fatLU = lu_solve(A,b)
print(fatLU)
```

Na Fatoração LU o vetor b é excluido se tornando vantajosa para a repetição do cálculo em um sistema igual com termos independentes diferentes. Já na Eliminação Gaussiana é vantajosa quando o novo sistema é mais fácil de resolver do que o anterior. Então, podemos observar que para a matriz que utilizamos a Fatoração LU é mais vantajosa baseado no tempo de execução.

# Métodos Iterativos

### Gauss-Jacobi

```python
def jacobi(A, b):
    
    inicio = time.time()
    tolerancia = 0.00000001
    iteracoes_maximas = 10000000
    x = np.zeros_like(b, dtype=np.double)
    iteracoes_jacobi = 0

    T = A - np.diag(np.diagonal(A))

    for iteracoes_jacobi in range(iteracoes_maximas):

        x_old = x.copy()

        x[:] = (b - np.dot(T, x)) / np.diagonal(A)

        if np.linalg.norm(x - x_old, ord=np.inf) / np.linalg.norm(x, ord=np.inf) < tolerancia:
            break

    fim = time.time()
    tempo_jacobi = fim - inicio

    return x, iteracoes_jacobi, tempo_jacobi

jacobi(A, b)
```

### Gauss-Seidel


```python
def seidel(A, b):
    
    inicio = time.time()
    tolerancia = 0.00000001
    iteracoes_maximas = 10000000
    x = np.zeros_like(b, dtype=np.double)
    iteracoes_seidel = 0

    for iteracoes_seidel in range(iteracoes_maximas):

        x_old = x.copy()

        for i in range(A.shape[0]):
            x[i] = (b[i] - np.dot(A[i, :i], x[:i]) - np.dot(A[i, (i + 1):], x_old[(i + 1):])) / A[i, i]

        if np.linalg.norm(x - x_old, ord=np.inf) / np.linalg.norm(x, ord=np.inf) < tolerancia:
            break

    fim = time.time()
    tempo_seidel = fim - inicio

    return x, iteracoes_seidel, tempo_seidel


seidel(A, b)
```

### Comparação

```python
print("Tempo Jacobi:", jacobi(A,b)[2])
print("Tempo Seidel:", seidel(A,b)[2])

grafJacobi = list(range(1, (jacobi(A, b)[1] + 1)))
lenJacobi = grafJacobi
grafSeidel = list(range(1, (seidel(A, b)[1] + 1)))
lenSeidel = grafSeidel

fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(lenJacobi, grafJacobi, label='Jacobi', color='b')
ax.plot(lenSeidel, grafSeidel, label='Seidel', color = 'r')
legend = ax.legend(loc='lower right', fontsize='x-large')
plt.xlabel('Nº de iterações', fontsize = 15)
plt.ylabel('Nº de iterações', fontsize = 15)
plt.title('Comparação do Nº de iterações', fontsize = 15)
plt.grid(True)
plt.show()
```

No método de Gauss-Jacobi, o cálculo é feito com os valores da iteração anterior, já no Gauss-Seidel em uma mesma iteração os valores já calculados são reutilizados para calcular os próximos. Ou seja, podemos perceber que o método de Gauss-Seidel é mais eficiente que o método de Gauss_Jacobi, visto que necessita de menos iterações.
