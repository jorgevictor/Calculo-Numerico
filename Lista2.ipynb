```python
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
def euler( x0, y0, xf, deltax, f ):
  n = int((xf-x0)/deltax)
  
  x = np.linspace(x0,xf,n)
  py = np.zeros([n])

  y = np.zeros([n])
  y[0] = y0

  for i in range(1,n):
    y[i] = deltax*f(x[i-1],y0) + y0
    y0 = y[i]

  #print("x_n\t    y_n")
  #for i in range(n):
  #  print(x[i],"\t",format(y[i],'6f'))

  return (x,y)
```

```python
def rk2( x0, y0, xf, deltax, f ):

  n = int((xf-x0)/deltax)

  x = np.linspace(x0,xf,n)
  py = np.zeros([n])

  y = np.zeros([n])
  y[0] = y0

  for i in range(1,n):
    py[i] = deltax*f(x[i-1],y[i-1]) + y[i-1]
    y[i] = deltax/2*( f(x[i],py[i]) + f(x[i-1],y[i-1]) ) + y[i-1]
  
  #print("x_n\t   py_n\t           y_n")
  #for i in range(n):
  #  print (x[i],"\t",format(py[i],'6f'),"\t",format(y[i],'6f'))

  return (x,y)
```

```python
def rk4( x0, y0, xf, deltax, f ):
  n = int((xf-x0)/deltax)

  x = np.linspace(x0,xf,n)
  py = np.zeros([n])

  y = np.zeros([n])
  y[0] = y0

  for i in range(1,n):
    k1 = deltax*f(x[i-1],y0)
    k2 = deltax*f(x[i-1]+deltax/2,y0+k1/2)
    k3 = deltax*f(x[i-1]+deltax/2,y0+k2/2)
    k4 = deltax*f(x[i-1]+deltax,y0+k3)
    y[i] =  y0 + (k1 + 2*k2 + 2*k3 + k4)/6
    y0 = y[i]

  print("x_n\t    y_n")
  for i in range(n):
    print(x[i],"\t",format(y[i],'6f'))
  return (x,y)
```

```python
def euler_2( x0, y0, xf, deltax, f ):
  n = int((xf-x0)/deltax)
  
  x = np.linspace(x0,xf,n)
  py = np.zeros([n])

  y = np.zeros([n])
  y[0] = y0
  erroRelativo = [abs(T(x0) - y0 ) /T(y0)]

  for i in range(1,n):
    y[i] = deltax*f(x[i-1],y0) + y0
    y0 = y[i]
    x0 = x0 + deltax
    erroRelativo.append(abs(T(x0) - y0 ) /T(y0))

  return  x, y, erroRelativo
```

```python
def rk2_2( x0, y0, xf, deltax, f ):

  n = int((xf-x0)/deltax)

  x = np.linspace(x0,xf,n)
  py = np.zeros([n])

  y = np.zeros([n])
  y[0] = y0
  erroRelativo = [abs(T(x0) - y0 ) /T(y0)]

  for i in range(1,n):
    py[i] = deltax*f(x[i-1],y[i-1]) + y[i-1]
    y[i] = deltax/2*( f(x[i],py[i]) + f(x[i-1],y[i-1]) ) + y[i-1]
    x0 = x0 + deltax
    erroRelativo.append(abs(T(x0) - y0 ) /T(y0))

  return x,y, erroRelativo
```

```python
def rk4_2( x0, y0, xf, deltax, f ):
  n = int((xf-x0)/deltax)

  x = np.linspace(x0,xf,n)
  py = np.zeros([n])

  y = np.zeros([n])
  y[0] = y0
  erroRelativo = [abs(T(x0) - y0 ) /T(y0)]

  for i in range(1,n):
    k1 = deltax*f(x[i-1],y0)
    k2 = deltax*f(x[i-1]+deltax/2,y0+k1/2)
    k3 = deltax*f(x[i-1]+deltax/2,y0+k2/2)
    k4 = deltax*f(x[i-1]+deltax,y0+k3)
    y[i] =  y0 + (k1 + 2*k2 + 2*k3 + k4)/6
    y0 = y[i]
    x0 = x0 + deltax
    erroRelativo.append(abs(T(x0) - y0 ) /T(y0))

  return x,y, erroRelativo
```

Questão 3.a)Implemente o Método de Euler para sistemas e resolva o sistema de Lotka-Volterra considerando a = 1.0, c = 0.75, α = γ = 0.5. Rode sempre 100/h iterações, dessa forma suas simulações terão sempre 100 unidades de tempo (dias, semanas, meses, anos... isso não importa).

```python
def derivada_x(x,y):
  a = 1.0
  alpha = 0.5
  
  return a*x - alpha*x*y

def derivada_y(x,y):
  c = 0.75
  gamma = 0.5
  return - c*y + gamma*x*y

def LotkaVolterra_EEuler(x_n, y_n, h):

 t = []
 tempo = 0
 x = []
 y = []

 for n in range(int(100/h)):
   x_n_mais_1  = x_n + h * derivada_x(x_n, y_n)
   y_n_mais_1 = y_n + h * derivada_y(x_n, y_n)
   tempo += h
   t.append(tempo)

   if(x_n_mais_1 < 0):
    x_n_mais_1 = 0

   if(y_n_mais_1 < 0 ):
    y_n_mais_1 = 0


   x.append(x_n_mais_1)
   y.append(y_n_mais_1)

   x_n = x_n_mais_1
   y_n = y_n_mais_1


 print(x[-1],y[-1]) 
 return x,y,t
```
